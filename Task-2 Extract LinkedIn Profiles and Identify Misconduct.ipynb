```python
import os
import re
import pandas as pd
```

```python
parent_folder = r"C:\Users\THOTA CHANDRASEKHAR\Downloads\Intership\GMT20240125-121850_RecordingnewChat\GMT20240125-121850_RecordingnewChat"

subfolders = []
text_files = []
file_content = []
for i in os.scandir(parent_folder):
    if i.is_dir():
        subfolders.append(i.path)
    for j in os.scandir(i):
        if j.is_file() and j.name.endswith('.txt'):
            text_files.append(j.path)
for k in text_files:
    with open(k, 'r', encoding='utf-8', errors='ignore') as file:
            
        file_content.append(file.read())
```

```python
file_content
```

```python
data=[]
for i in file_content:
    ni=i.splitlines()
    for j in ni:
        nj=j.split("\t")
        data.append(nj)
```

```python
data
```

```python
df=pd.DataFrame(data)
```

```python
df.columns=("Time","Name","Message")
```

## Out of all the LinkedIn URLs shared over chat, how many are valid profile URLs

```python
valid_links = []
for i in df1['Message']:
    vl = re.findall(r'\b(?:https?://)?(?:www.)?linkedin.com/in/[a-zA-Z0-9-]+/?\b', i)
    if vl:
        valid_links.append(vl)
    else:
        pass
```
